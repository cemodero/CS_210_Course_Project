```python
import pandas as pd

# Replace 'path_to_your_csv_file.csv' with the actual path to your CSV file
file_path = 'daylio_export_new_2024_01_17.csv'

# Loading the CSV file into a DataFrame
mood_df = pd.read_csv(file_path)

# Displaying the first few rows of the DataFrame
print(mood_df.head())
```

```python
# Dropping the 'note_title' column
mood_data = mood_df.drop(columns=['note_title'])

mood_data.head(5)
```

```python
# Fill null values in the 'note' column with 'Neutral' or 'Normal'
mood_data['note'].fillna('Neutral', inplace=True)

# Display the first few rows of the DataFrame after filling null values
mood_data.head(5)
```

```python
# Fill null values in the 'activities' column with 'No Activity'
mood_data['activities'].fillna('No Activity', inplace=True)

# Display the first few rows of the DataFrame after filling null values
mood_data.info()
```

```python
# Fill the rows in the 'note' column with 'Neutral' if they are not 'On track', 'Relapsing', or 'Unstable'
mood_data['note'] = mood_data['note'].apply(lambda x: 'Neutral' if x not in ['On track', 'Relapsing', 'Unstable'] else x)

# Display the first few rows of the DataFrame after modification
mood_data.head()
```

```python
# Reverse the order of the DataFrame
mood_data_reversed = mood_data.iloc[::-1].reset_index(drop=True)

# Display the first few rows of the reversed DataFrame
mood_data_reversed.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import random

# Assuming mood_data_reversed_filtered and filtered_streaming_df are already loaded
# Merge the mood data and streaming data
combined_df = pd.merge(filtered_streaming_df, mood_data_reversed_filtered, left_on='ts_date', right_on='full_date')

# Handle different types of activity entries
def process_activities(activity):
    if activity == 'No Activity':
        return ['No Activity']
    elif '|' in activity:
        return [act.strip() for act in activity.split('|')]
    else:
        return [activity.strip()]

combined_df['activities_list'] = combined_df['activities'].apply(process_activities)

# Mood encoding dictionary (from worst to best)
mood_encoding = {'awful': 0, 'bad': 1, 'meh': 2, 'good': 3, 'rad': 4}

# One-hot encode activities
mlb = MultiLabelBinarizer()
activities_encoded = mlb.fit_transform(combined_df['activities_list'])
activities_df = pd.DataFrame(activities_encoded, columns=mlb.classes_)

# Apply the ordered mood encoding
combined_df['mood_encoded'] = combined_df['mood'].map(mood_encoding)

# Prepare features and target
X = pd.concat([combined_df[['mood_encoded']], activities_df], axis=1)
y = combined_df['master_metadata_track_name']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model training
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Example prediction with variability
def predict_song(mood, activities):
    mood_encoded = mood_encoding[mood]
    activities_df = pd.DataFrame(mlb.transform([activities]), columns=mlb.classes_)
    features = pd.concat([pd.DataFrame({'mood_encoded': [mood_encoded]}), activities_df], axis=1)
    features = features.reindex(columns=X_train.columns, fill_value=0)

    # Get prediction probabilities
    probas = model.predict_proba(features)[0]

    # Get a list of classes (song names) and their probabilities
    classes = model.classes_
    class_probas = list(zip(classes, probas))

    # Filter out zero-probability classes to avoid unnecessary computation
    class_probas = [(song, prob) for song, prob in class_probas if prob > 0]

    # Randomly select a song based on probabilities
    songs, probs = zip(*class_probas)
    chosen_song = random.choices(songs, weights=probs, k=1)[0]

    return chosen_song

# Test the function
example_mood = 'rad'
example_activities = ['friends']
recommended_song = predict_song(example_mood, example_activities)
print(f"Recommended track for mood '{example_mood}' and activities '{example_activities}': {recommended_song}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample moods and activities
sample_moods = ['awful', 'good', 'meh']
sample_activities = ['family', 'reading', 'movies']

# Generate recommendations
recommendations = []
for mood in sample_moods:
    for activity in sample_activities:
        for _ in range(10):  # Generate 10 recommendations per combination
            song = predict_song(mood, [activity])
            recommendations.append({'Mood': mood, 'Activity': activity, 'Song': song})

# Create DataFrame
recommendation_df = pd.DataFrame(recommendations)

# Count frequency of song recommendations
heatmap_data = recommendation_df.groupby(['Mood', 'Activity', 'Song']).size().reset_index(name='Count')

# Pivot for heatmap
heatmap_data_pivot = heatmap_data.pivot_table(index=['Mood', 'Activity'], columns='Song', values='Count', fill_value=0)

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data_pivot, annot=True, cmap='viridis')
plt.title('Song Recommendations Frequency by Mood and Activity')
plt.xlabel('Song')
plt.ylabel('Mood and Activity')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import random
from collections import Counter

# Load data for version 1
# Assuming mood_data_reversed_filtered and filtered_streaming_df are already loaded

# Merge the mood data and streaming data
combined_df_v1 = pd.merge(filtered_streaming_df, mood_data_reversed_filtered, left_on='ts_date', right_on='full_date')

def process_activities_v1(activity):
    if activity == 'No Activity':
        return ['No Activity']
    elif '|' in activity:
        return [act.strip() for act in activity.split('|')]
    else:
        return [activity.strip()]

combined_df_v1['activities_list'] = combined_df_v1['activities'].apply(process_activities_v1)

mood_encoding_v1 = {'awful': 0, 'bad': 1, 'meh': 2, 'good': 3, 'rad': 4}

mlb_v1 = MultiLabelBinarizer()
activities_encoded_v1 = mlb_v1.fit_transform(combined_df_v1['activities_list'])
activities_df_v1 = pd.DataFrame(activities_encoded_v1, columns=mlb_v1.classes_)

combined_df_v1['mood_encoded'] = combined_df_v1['mood'].map(mood_encoding_v1)

X_v1 = pd.concat([combined_df_v1[['mood_encoded']], activities_df_v1], axis=1)
y_v1 = combined_df_v1['master_metadata_track_name']

X_train_v1, X_test_v1, y_train_v1, y_test_v1 = train_test_split(X_v1, y_v1, test_size=0.2)

model_v1 = RandomForestClassifier()
model_v1.fit(X_train_v1, y_train_v1)

def calculate_association_score_v1(combined_df):
    mood_activity_song_count_v1 = Counter()
    total_song_count_v1 = Counter()

    for _, row in combined_df.iterrows():
        mood = row['mood']
        activities = tuple(sorted(row['activities_list']))
        song = row['master_metadata_track_name']

        mood_activity_song_count_v1[(mood, activities, song)] += 1
        total_song_count_v1[song] += 1

    association_scores_v1 = {}
    for key, count in mood_activity_song_count_v1.items():
        mood, activities, song = key
        total_count = total_song_count_v1[song]
        score = count / total_count if total_count > 0 else 0
        association_scores_v1[(mood, activities, song)] = score

    return association_scores_v1

association_scores_v1 = calculate_association_score_v1(combined_df_v1)

def predict_song_v1(mood, activities):
    mood_encoded = mood_encoding_v1[mood]
    activities_df = pd.DataFrame(mlb_v1.transform([activities]), columns=mlb_v1.classes_)
    features = pd.concat([pd.DataFrame({'mood_encoded': [mood_encoded]}), activities_df], axis=1)
    features = features.reindex(columns=X_train_v1.columns, fill_value=0)

    probas = model_v1.predict_proba(features)[0]

    classes = model_v1.classes_

    adjusted_probas = []
    for class_idx, song in enumerate(classes):
        score_key = (mood, tuple(sorted(activities)), song)
        association_score = association_scores_v1.get(score_key, 0)
        adjusted_prob = probas[class_idx] * association_score
        adjusted_probas.append(adjusted_prob)

    total = sum(adjusted_probas)
    normalized_adjusted_probas = [prob / total for prob in adjusted_probas] if total > 0 else [1 / len(classes)] * len(classes)

    chosen_song = random.choices(classes, weights=normalized_adjusted_probas, k=1)[0]

    return chosen_song

# Test the function
example_mood = 'bad'
example_activities = ['studying']
recommended_song = predict_song_v1(example_mood, example_activities)
print(f"Recommended track for mood '{example_mood}' and activities '{example_activities}': {recommended_song}")
```
